```python
import numpy as np  # Se importa la librería NumPy, la cual permite trabajar con arreglos y estructuras
                    # matemáticas de manera eficiente. Es fundamental en Machine Learning
                    # y procesamiento de datos.

# Ejercicio 1: Dimensiones y forma de arreglos

# Se crea un vector unidimensional (arreglo de una sola dimensión)
vector = np.array([1,2,3,4,5])

# Se crea una matriz bidimensional (2 filas y 3 columnas)
matriz = np.array([[1,2,3],
                   [4,5,6]])

# vector.ndim devuelve el número de dimensiones del arreglo
# vector.shape devuelve la forma del arreglo (tamaño por dimensión)
print(vector.ndim, vector.shape)

# matriz.ndim devuelve el número de dimensiones (en este caso 2)
# matriz.shape devuelve la forma (filas, columnas)
print(matriz.ndim, matriz.shape)
```

```python
import numpy as np  # Se importa NumPy para realizar operaciones matemáticas con arreglos

# Ejercicio 2: Operaciones básicas con matrices

# Se define la matriz A de 2x2
A = np.array([[1,2],
              [3,4]])

# Se define la matriz B de 2x2
B = np.array([[5,6],
              [7,8]])

# Se realiza la suma elemento a elemento entre A y B
# NumPy permite hacer esta operación directamente sin usar ciclos
C = A + B

# Se calcula la raíz cuadrada de cada elemento de la matriz A
# np.sqrt() aplica la operación a cada elemento individualmente
D = np.sqrt(A)

# Se imprimen los resultados
print(C)
print(D)
```

```python
import numpy as np  # Librería fundamental para manejar arreglos numéricos

# Ejercicio 3: Simulación y procesamiento básico de una imagen

# Se genera una "imagen" simulada de 5x5 píxeles
# Los valores van de 0 a 255, que es el rango típico en imágenes en escala de grises
imagen = np.random.randint(0, 256, (5,5))

# Región de Interés (ROI)
# Se extrae una región específica de la imagen.
# imagen[1:5, 1:5] significa:
# Filas desde la posición 1 hasta la 4
# Columnas desde la posición 1 hasta la 4
# Esto se conoce como slicing en NumPy
roi = imagen[1:5, 1:5]

# Normalización
# Se normalizan los valores de la imagen dividiendo entre 255
# Esto convierte los valores del rango [0,255] a [0,1]
# La normalización es importante en Machine Learning
# porque mejora la estabilidad y el rendimiento de los modelos
imagen_norm = imagen / 255.0

# Se imprimen los resultados
print(roi)
print(imagen_norm)
print(imagen)
```

```python
# Ejercicio 1: Figuras y Subplots

import numpy as np
import matplotlib.pyplot as plt

# Se genera un arreglo de 100 valores entre 0 y 2π
# linspace permite crear datos continuos, útiles para graficar funciones
x = np.linspace(0, 2*np.pi, 100)

# Se crea una figura con tamaño personalizado (ancho=8, alto=4)
plt.figure(figsize=(8,4))

# Primer subplot (1 fila, 2 columnas, posición 1)
plt.subplot(1,2,1)
plt.plot(x, np.sin(x))
plt.title("Función Seno")

# Segundo subplot (1 fila, 2 columnas, posición 2)
plt.subplot(1,2,2)
plt.plot(x, np.cos(x))
plt.title("Función Coseno")

# Se muestran ambas gráficas
plt.show()
```

```python
# Ejercicio 2: Gestión de Gráficas

import numpy as np
import matplotlib.pyplot as plt

# Se generan 100 valores entre -10 y 10
# Esto permite analizar el comportamiento de la función en un rango amplio
x = np.linspace(-10, 10, 100)

# Se define la función cuadrática y = x^2
# Esta operación se realiza de forma vectorizada (sin ciclos)
y = x**2

# Se crea una nueva figura
plt.figure()

# Se grafica la función y se agrega una etiqueta para la leyenda
plt.plot(x, y, label="y = x**2")

# Se agregan elementos descriptivos a la gráfica
plt.title("Funcion Cuadratico")
plt.xlabel("X")
plt.ylabel("y")

# Se muestra la leyenda usando el label definido anteriormente
plt.legend()

# Se activa la cuadrícula para facilitar la interpretación visual
plt.grid(True)

# Se muestra la gráfica final
plt.show()
```

```python
# Ejercicio 1: Series y DataFrames

import pandas as pd  # Pandas se utiliza para manipulación y análisis de datos estructurados

# Se crea un diccionario que simula datos de píxeles
# Cada clave representa una característica (feature)
# y cada lista representa diferentes muestras
datos = {
    "pixel_1": [120, 130, 125],
    "pixel_2": [60, 65, 70],
    "clase": [0, 1, 0]  # Etiqueta o clase asociada a cada muestra
}

# Se convierte el diccionario en un DataFrame
# Un DataFrame es una estructura tabular similar a una tabla de base de datos
df = pd.DataFrame(datos)

# Se imprime el contenido del DataFrame
print(df)

# df.shape devuelve (filas, columnas)
# Esto permite conocer cuántas muestras y características tenemos
print(df.shape)
```

```python
# Ejercicio 2: Indexado y Filtrado

# Se filtran únicamente las filas donde la columna "clase" sea igual a 1
# Esta operación se conoce como filtrado booleano en Pandas
# Es muy utilizada en Machine Learning para separar clases o subconjuntos de datos
df_clase_1 = df[df["clase"] == 1]

# Selección de columnas específicas

# Se seleccionan únicamente las columnas "pixel_1" y "pixel_2"
# df.loc permite acceder a filas y columnas mediante etiquetas
# ":" indica que se seleccionan todas las filas
pixeles = df.loc[:, ["pixel_1", "pixel_2"]]

# Se imprimen los resultados
print(df_clase_1)
print(pixeles)
```

```python
# Ejercicio 3: Aritmética y Manipulación Algebraica

import pandas as pd  # Librería utilizada para manipulación y análisis de datos estructurados

# Se crea un diccionario con valores simulados de píxeles
# Cada fila representa una muestra y cada columna una característica
datos = {
    "pixel_1": [120, 130, 125],
    "pixel_2": [60, 65, 70],
    "clase": [0, 1, 0]
}

# Se convierte el diccionario en un DataFrame
df = pd.DataFrame(datos)

# Se seleccionan únicamente las columnas que contienen características (features)
# Se excluye la columna clase porque normalmente es la variable objetivo
pixeles = df.loc[:, ["pixel_1", "pixel_2"]]

# Normalización simple

# Se dividen los valores entre 255
# Esto escala los datos al rango aproximado [0,1]
# Es un paso común en visión por computadora para estabilizar el entrenamiento
pixeles_norm = pixeles / 255.00

# Media por columna

# Se calcula el promedio de cada columna
# Esto permite conocer la intensidad promedio de cada pixel en el conjunto de datos
media = pixeles.mean()

print(pixeles_norm)
print(media)
```
